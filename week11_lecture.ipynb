# Section 1: Dictionaries

## Why dictionaries?

Last few weeks we used **lists** to store many items in order. But lists only store one piece of information per slot. What if you want to store *several pieces of information about the same thing*?

````python
# The clunky old way: parallel lists
names = ["Alice", "Bob", "Charlie"]
ages = [20, 19, 21]
majors = ["CS", "Math", "CS"]

# To get Bob's info you find his index, then use it on every list
index = names.index("Bob")
print(f"{names[index]}, age {ages[index]}, major {majors[index]}")

# Three sources of truth that have to stay in sync. Easy to mess up.
````

## Creating and accessing

A **dictionary** stores **key-value pairs** inside curly braces `{}`. The key is a label; the value is the data. You read it back with `dict[key]`.

````python
# Same data, but bundled together
student = {
    "name": "Bob",
    "age": 19,
    "major": "Math"
}

print(student)
print(f"Name: {student['name']}")
print(f"Age:  {student['age']}")
````

## Adding and updating

Assigning to a key that does not exist **creates** it. Assigning to a key that exists **updates** it. Same syntax, different result.

````python
student = {"name": "Bob", "age": 19}
print("Start: ", student)

student["age"] = 20         # update
student["major"] = "Math"   # add new key
print("After: ", student)
````

## Safe lookup with `.get()`

Reading a key that does not exist will **crash** with `KeyError`. If you are not 100 percent sure the key is there, use `.get(key, default)`. It returns the default instead of crashing.

````python
student = {"name": "Eve", "age": 22}

# Safe: returns the default "unknown" instead of crashing
major = student.get("major", "unknown")
print(f"Major: {major}")

# Also safe: check if a key exists with `in`
if "email" in student:
    print("has email")
else:
    print("no email on file")
````

## Looping through a dictionary

Most useful pattern: `.items()` gives you each key and value at once. You unpack them into two loop variables.

````python
grades = {"Alice": 92, "Bob": 85, "Charlie": 78, "Diana": 95}

# Print each pair
for name, score in grades.items():
    print(f"{name} earned {score}")

print()

# Compute an average using just the values
total = 0
for score in grades.values():
    total += score
average = total / len(grades)
print(f"Class average: {average:.1f}")
````

# Section 2: Functions Pt 1

## Why functions?

You have used functions all semester (`print()`, `len()`, `input()`, `range()`). Tonight you start writing your own. The reason: **DRY - Don't Repeat Yourself**. If a chunk of logic shows up in three places, give it a name and call it three times instead.

````python
# Without a function: same shape repeated three times
print("-" * 25)
print("Hello, Alice. Welcome.")
print("-" * 25)

print("-" * 25)
print("Hello, Bob. Welcome.")
print("-" * 25)

print("-" * 25)
print("Hello, Charlie. Welcome.")
print("-" * 25)
# If we want to change the wording, we have to change it in three places.
````

## Defining and calling

Use `def` to **define** a function. Use `name()` to **call** it.

````python
# Define once
def greet(name):
    print("-" * 25)
    print(f"Hello, {name}. Welcome.")
    print("-" * 25)

# Call as many times as you want
greet("Alice")
greet("Bob")
greet("Charlie")
````

**Vocabulary:**
- **Parameter** - the variable name in the definition (`def greet(name):` - `name` is the parameter)
- **Argument** - the actual value you pass in when calling (`greet("Alice")` - `"Alice"` is the argument)

They go together. The parameter receives whatever the argument is.

## Multiple parameters

````python
def area(length, width):
    result = length * width
    print(f"A {length} x {width} rectangle has area {result}")

area(10, 5)
area(7, 3)
area(15, 8)
````

## `return` vs `print` (this trips everyone up)

**`print`** shows a value on the screen. The caller cannot do anything with what was printed.

**`return`** hands a value back to the line that called the function. The caller can store it, use it in math, pass it to another function, etc.

Rule: if anyone needs the result later, **return** it.

````python
# Version A: prints. Caller cannot reuse the result.
def area_print(length, width):
    print(length * width)

area_print(10, 5)            # screen shows 50
# but I can't store 50 anywhere
````

````python
# Version B: returns. Caller can store and reuse.
def area(length, width):
    return length * width

room1 = area(10, 5)
room2 = area(8, 6)
total = room1 + room2
print(f"Two rooms total {total} square feet")
````

# Section 3: try / except basics

## The crash problem

Reading user input is risky. `input()` always returns a **string**. If you want a number, you have to convert with `int()` or `float()`. If the user types something that is not a number, the conversion **fails** and the whole program ends with a `ValueError`.

*This is encouraged in your final project, not required.* But it is a small change for a big win.

````python
# This works fine when the user cooperates:
# age = int(input("Enter your age: "))

# But if they type "twenty", you get:
#   ValueError: invalid literal for int() with base 10: 'twenty'
# and the program crashes.

# Demo without input prompt so the notebook can run:
user_typed = "twenty"
# int(user_typed)   # this would raise ValueError
print("Without try/except, a bad input ends the whole program.")
````

## The basic pattern

```python
try:
    # the line that might crash
except:
    # what to do if it does
```

If the `try` block succeeds, the `except` block is skipped. If it fails, Python jumps to `except` instead of crashing.

````python
user_typed = "twenty"

try:
    age = int(user_typed)
    print(f"Got age {age}")
except ValueError:
    print("That was not a number. Please try again.")

print("Program keeps running either way.")
````

## A more useful pattern: keep asking until you get something valid

This combines a `while` loop with `try / except`. It is the standard recipe for input validation, and it is exactly what you would use in a final project that asks for a number.

````python
# Demo with a list of fake user inputs so it runs without prompting.
# In your project you would use input() inside the loop.
fake_inputs = ["twenty", "abc", "25"]
attempt = 0

while True:
    typed = fake_inputs[attempt]
    attempt += 1
    try:
        age = int(typed)
        print(f"Got age {age}")
        break              # success: leave the loop
    except ValueError:
        print(f"'{typed}' is not a number. Try again.")

print("Moving on with age =", age)
````

Pseudocode is a plain-English plan for your program, one line per step. It is not Python and it does not have to run. It just has to be **specific enough that you can sit down later and translate it line by line into code.**

It is **not** a paragraph describing what your program will do ("My program will be a quiz that asks the user questions...") - that is a project description, not pseudocode. Every required piece (dictionary, loop, function, input, output) needs to appear as concrete, ordered steps.

## A worked example: order calculator

Goal: a small program that lets the user pick items from a coffee menu, totals the order, and prints a receipt.

### The pseudocode (this is what you submit)

```
Create a dictionary called menu with item names as keys and prices as values
    "latte" maps to 4.50
    "espresso" maps to 3.00
    "cold brew" maps to 5.00

Create an empty list called order_items
Display "Welcome to the Coffee Order System"

Loop

    If the typed item is in menu

    Otherwise
        Display "Sorry, we don't have that item"

Define a function calculate_total that takes order_items and menu

    For each item in order_items

    Return subtotal

Call calculate_total with order_items and menu, store result in total
Display each ordered item and the total formatted to two decimals
Display "Thank you for your order"
```

Notice: no colons, no Python keywords like `def` or `for ... in ... :`, no brackets. Plain action verbs (Create, Display, Ask, Loop, Define, For each, Otherwise) and indentation showing what is inside the loop and inside the function.

## Pseudocode tips

1. **Plain English, not Python.** No colons after `For each` or `If`. No `print()`. Use action verbs: Ask, Display, Set, Add, Loop, Call, Return.
2. **Indent to show structure.** Body of a loop, body of a function, contents of a try block - all indented under their header.
3. **Use real names.** "Define a function `calculate_total`" not "Define a function thing_doer".
4. **One step per line.** If a line is doing two things, split it.
